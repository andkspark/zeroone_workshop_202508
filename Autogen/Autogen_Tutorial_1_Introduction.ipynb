### AI Agent 소개

“날씨가 따뜻한 곳으로 여행 가고 싶어~” 한마디에 여행지 검색, 최저가 비행기 및 숙소 예약까지 착착 해내고, 옷차림까지 조언해주는 AI 비서를 곁에 둘 날이 있을까요?

자동차가 날아다니는 먼 미래의 공상과학기술 일 것 같지만, 사실 그렇게 멀지 않은 미래에 실현될 기술이랍니다. 바로 AI Agent 덕분입니다!

### AI Agent란?

  전통적인 강화학습론에서는 Agent를 “환경”을 “인식”하고 그에 따라 “행동” 하는 개체로 정의합니다.

  알림을 설정하고, 날씨를 확인하고, 좋아하는 음악을 틀어주는 핸드폰의 디지털 비서가 좋은 전통적인 Agent의 예시입니다. 주로 단순 작업 자동화에 초점을 맞추었으며, 제한된 자율성과 의사결정 능력을 가졌습니다. 이러한 Agent들은 본질적으로 사용자의 지시에 따라 정해진 범위 내에서 작동합니다.

  그러다 LLM의 발전에 맞추어 언어 모델을 탑재하면서, Agent system은 사용자와 더 자연스러운 대화가 가능해지고, 방대한 정보를 처리하면서도 다양한 환경과 상호작용할 수 있는 형태로 발전하였습니다. Agent 기술은 고객 서비스, 의료, 금융, 물류 등 다양한 산업 분야에 도입되기 시작했습니다.

### Agentic AI
Capture.png


2024년에 들어 LLM을 탑재한 AI Agent 가 급부상하기 시작하면서, 연구자들은 서로 조금씩 다르게 AI Agent를 정의하고 있습니다만, 다음의 특징을 가지고 있을 때 **Agentic** 하다고 공통적으로 정의합니다.

- AI 시스템이 복잡한 환경에서 사용자 지시 없이 어려운 목표를 추구할 수 있는 경우
- 자연어 지시 하에 사용자의 인터페이스나 감독 없이 자율적으로 행동할 수 있는 경우
- 웹 검색, 프로그래밍 등의 도구를 사용할 수 있거나 계획할 수 있는 경우

위에 언급한 Agentic 하다는 단어가 기존의 Agent 시스템 간의 차이를 잘 설명할 수 있는 단어가 될 것 같습니다.

image.webp

그렇다면 Agentic AI 와 AI Agent 로 칭할 수 있는 것들은 무엇이 있을까요?

- 환경에 따라 주행을 조정하는 자율주행 자동차
- 의료 데이터를 분석하고, 패턴을 식별하며, 의사가 더 정보에 입각한 결정을 내릴 수 있도록 돕는 헬스케어 시스템

등 복잡하고 상황이 변하는 것은 Agentic AI 로,

- 알림을 설정하고, 날씨를 확인하고, 좋아하는 음악을 틀어주는 개인 비서 앱
- 코드 작성을 돕는 Copilot

등 정의된 명령에 의존하며 간단하고 반복적인 작업을 처리하는 것들은 AI Agent 로 구분할 수 있습니다.

### **Agentic AI 를 구성하는 것들**

그렇다면, 사용자가 원하는 목표를 이루기 위해 어떤 것들이 있을까요?

image (1).webp

2025년 구글은 Agent white paper를 통해 Agent system 의 핵심 구성 요소를 다음과 같이 정의하였습니다.

- 모델 (Model)
    - 모델(Model)은 에이전트(Agent)의 의사 결정을 담당하는 핵심 요소입니다. 텍스트, 이미지 생성 등 다양한 작업을 수행할 수 있는 대규모 멀티모달 모델(LMM)을 활용합니다. ReAct, Chain-of-Thought, Tree-of-Thoughts 등의 프레임워크를 따라 복잡한 추론을 수행합니다.
- 도구 (Tools)
    - 에이전트(Agent)가 외부 데이터 및 서비스와 상호작용할 수 있도록 지원하는 요소입니다. API 호출, 데이터베이스 검색, 실시간 정보 접근 등 다양한 형태로 존재하며, 에이전트(Agent)가 실제 세계와 연결되는 핵심 통로 역할을 합니다.
- .오케스트레이션 레이어 (Orchestration Layer)
    - 오케스트레이션 레이어는 에이전트가 정보를 수집하고, 추론을 수행하며, 최종 결정을 내리는 순환적인 프로세스를 관리합니다. 이 레이어는 메모리와 상태를 유지하며, 다양한 프레임워크를 사용해 복잡한 작업을 수행합니다.

모델과 Agent의 차이점을 더 명확하게 하기 위해, 중요한 부분을 짚고 넘어가겠습니다

### Agentic AI 가 작동하는 방식

그렇다면 위의 구성 요소로 구성된 Agentic AI system 은 어떻게 동작할까요 ?

e3daae3d-aaff-4caf-9f8b-055ce21aae33.webp

요리사가 복잡한 요리를 준비하는 과정에 비유하여, Agent의 작동 방식을 설명할 수 있습니다!

1단계: 정보 수집

- 요리사는 고객 주문에 맞춰 식재료를 수집하며 요리를 준비합니다.
- 에이전트는 사용자 쿼리를 입력받거나 외부 데이터를 조회합니다.

2단계: 내부 추론

- 요리사는 수집된 정보를 바탕으로 어떤 요리를 만들지 결정합니다.
- 에이전트는 모델을 활용하여 다음 행동을 결정합니다.

3단계: 행동 실행

- 요리사는 재료를 손질하고 요리를 만듭니다.
- 에이전트는 도구를 사용하여 외부 시스템과 상호작용합니다.

4단계: 피드백 및 조정

- 요리사가 요리 중간에 맛을 보고, 부족한 부분을 추가하거나 요리 순서를 변경합니다.
- 에이전트는 오케스트레이션 레이어에서 이전 결과에 기반하여 다음 행동을 조정합니다.

위의 과정을 통해, 요리사와 에이전트는 과정을 반복하며 점진적으로 목표를 달성하게 됩니다.

사용자는 이러한 에이전트의 동작을 오케스트레이션 레이어에서 메모리, 상태, 추론, 계획을 관리할 수 있습니다.

### Agentic AI의 전망

2023년, 빌 게이츠는 향후 5년 이내에 단일 AI 에이전트가 사용자의 음성 명령을 이해하고 이를 바탕으로 다양한 작업을 수행함으로써, 더 이상 여러 앱을 각각 사용할 필요가 없어질 것이라고 전망하였습니다. 이와 같은 에이전트 기반의 AI 시스템은 일상 언어로 문서 작성, 스프레드시트 작성, 이메일 발송 등의 컴퓨터 작업을 일괄적으로 처리할 수 있게 하며, 이는 기존의 CLI에서 GUI로의 전환을 뛰어넘는 인터페이스 혁명에 비하는 큰 변화를 의미합니다.

1111.png

이런 흐름은 2024년 12월 개최된 NeurIPS 2024에서도 찾아볼 수 있었는데요! AI 에이전트를 주제로 한 106편의 논문이 제출되었으며, MIT Media Lab의 MD Agent는 의료 현장의 협업 구조를 반영해, 작업의 성격에 따라 단일 또는 다중 LLM 에이전트를 유연하게 배치하는 방식을 제안했습니다. 이를 통해 기존 방식보다 더 정교한 의사결정 흐름을 구현하였습니다.

같은 행사 내 Expo 세션에서는 카드사 Capital One이 다중 에이전트를 활용한 고객 상담 시스템을 운용하고 있음을 밝혔으며, emergence.ai는 에이전트 오케스트레이션을 이용해 여행 일정 생성부터 항공권, 숙소 예약까지 자동화하는 실제 시연을 통해 응용 가능성을 보여주었습니다.

2025년에 들어, 많은 연구자들이 다양한 분야에 적용가능한 멀티에이전트 시스템을 활용한 프레임워크를 제안하고 있으며, 연구 분야의 지원을 위한 다양한 오케스트레이션 툴 또한 경쟁적으로 소개되고 있습니다.

### LangGraph

LangGraph는 LangChain에서 제공하는 에이전트 구축 프레임워크로, 복잡한 작업을 처리하는 확장 가능한 에이전트 설계를 지원합니다. 특히 단일 에이전트, 멀티 에이전트, 계층적, 순차적 실행 등 다양한 제어 흐름을 제공합니다. 또한, 상태 저장 기능으로 장기 컨텍스트를 유지하고 실시간 스트리밍으로 사용자 경험을 향상시킵니다.
